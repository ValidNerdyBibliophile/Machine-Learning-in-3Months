```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img.jpg")
```

```python
img.shape #3 is the BGR count
```

```python
# convert the image to grayscale

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "gray")
```

```python
U,S,V = np.linalg.svd(gray,full_matrices = False)
```

```python
U. shape
```

```python
S.shape
```

```python
V.shape
```

```python
comps = [U.shape[0],100,80,50,35,20,5,2]
plt.figure(figsize = (20,15))
for i in range(len(comps)):
    trans_img = U[:,:comps[i]]@np.diag(S[:comps[i]])@ V[:comps[i],:]
    plt.subplot(3,4,i+1)
    plt.imshow(trans_img,cmap = "gray")
    
    plt.title("Image with n_component "+str([comps[i]]))
```
