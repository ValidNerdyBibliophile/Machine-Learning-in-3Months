### Uniform Distribution

Perhaps one of the simplest and useful distribution is the uniform distribution. 

```python
from scipy.stats import uniform
n = 100000
start = 10
width = 20
data_uniform = uniform.rvs(size = n, loc = start, scale = width)
ax = sns.displot(data_uniform,bins = 10,  kde = True, color = "skyblue")
ax.set(xlabel = "Uniform Distribution", ylabel = "Frequency")
```

### Normal Distribution

```python
from scipy.stats import norm
data_normal = norm.rvs(size = 20000,loc = 0,scale = 1)
ax = sns.histplot(data_normal, kde = True, color = "violet")
ax.set(xlabel = "Normal Distribution", ylabel = "Frequency")
```

### Gamma Distribution

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(size = 20000, a =5)
ax = sns.histplot(data_gamma, kde = True, color = "violet")
ax.set(xlabel = "Gamma Distribution", ylabel = "Frequency")
```

### Exponential Distribution

The exponential distribution describes the time between events in a Poisson point process ie a process in which events occur continuously and independently at a constant average rate.

```python
from scipy.stats import expon
data_expon = expon.rvs(size = 20000, loc = 0, scale = 1)
ax = sns.histplot(data_expon, bins = 100, kde = True, color = "k")
ax.set(xlabel = "Exponential Distribution", ylabel = "Frequency")
```

### Poisson Distribution

A number of times event occured in particular time interval

```python
from scipy.stats import poisson
data_poisson = poisson.rvs(size = 20000, mu = 5)
ax = sns.histplot(data_poisson, bins = 100, kde = True, color = "k")
ax.set(xlabel = "Poisson Distribution", ylabel = "Frequency")
```

### Binomial Distribution
only two outcomes are possible, such as success or faillure, gain or loss, win or lose

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10, p =0.8, size = 10000 )
ax = sns.histplot(data_binom, bins = 100, kde = False, color = "k")
ax.set(xlabel = "Binomial Distribution", ylabel = "Frequency")
```

### Bermouilli Distribution

1 - success
0 - failure

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(p =0.6, size = 10000 )
ax = sns.histplot(data_bern, kde = False, color = "k")
ax.set(xlabel = "Bernoulli Distribution", ylabel = "Frequency")
```
