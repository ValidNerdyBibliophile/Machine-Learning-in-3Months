### Lasso Regression
Converts the coefficient to ***ABSOLUTE ZERO*** . So it changes the accuracy as it affects the whole model.

```python
import pandas as pd
data = pd.read_csv("housing.csv")
```

```python
data.head()
```

```python
import seaborn as sns
sns.pairplot(data)
```

```python
x = data.drop(columns = ["MEDV"])
```

```python
y = data["MEDV"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_transform = PolynomialFeatures(degree = 3)
```

```python
x_poly = poly_transform.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
poly_model = LinearRegression()
```

```python
poly_model.fit(x_poly,y)
ypred = poly_model.predict(x_poly)
```

```python
pd.DataFrame({"Actual":y,"Predicted":ypred})
```

```python
poly_model.score(x_poly,y)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha =0.3)
ridge_model.fit(x,y)
```

```python
ridge_model.score(x,y)
```

```python
ridge_model.coef_
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 0.6)
```

```python
lasso_model.fit(x,y)
```

```python
lasso_model.coef_
```

```python
lasso_model.score(x,y)
```
