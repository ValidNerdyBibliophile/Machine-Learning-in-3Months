### Ridge Regression

##### Used when overfitting issues arises. Eg: when training model gives 90%+ accuracy but testing model gives 50 or 60% accuracy!!

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha =0.3) #alpha is a regularization term. Irrelevant features impact is reduced a little.
ridge_model.fit(x,y)
```

```python
ridge_model.score(x,y)
```

```python
ridge_model.coef_  #slope changes a little when alpha increases.
```

```python
ridge_model.intercept_
```
