### Scatter Plot

```python
sns.scatterplot(x = 'sepal_length', y = "sepal_width", data = iris_df)
```

### Swarm Plot

When u have categorical data, we use swarm plot as it has the ability to categorize the data sperately, with 1 condition that: **_x_ must have the column name of the data which is categorical data.**

```python
sns.swarmplot(x = "species", y = "sepal_width", data = iris_df)
```

### Bar Plot

```python
sns.barplot(x ="sepal_length", y = "sepal_width", data = iris_df)
```

### Histogram

```python
sns.histplot(x = "sepal_width", data = iris_df)
```

### Line Plot

```python
sns.lineplot(x = "sepal_width", y = "sepal_length", data = iris_df)
```

###  Relational plot

```python
sns.relplot(x = "sepal_width", y= "sepal_length", data = iris_df)
```

### Regression Plot

```python
sns.regplot(x = "sepal_width", y= "sepal_length", data = iris_df)
```

### Joint Plot

```python
sns.jointplot(x = "sepal_width", y= "sepal_length", data = iris_df)
```

### Factor Plot/ Cat Plot

```python
sns.factorplot(x ="species", data = iris_df, kind = "count")
```

_can use catplot instead of factor plot. Cause in future factor plot is going to be renamed as catplot_

```python
sns.catplot(x ="species", data = iris_df, kind = "count")
```

### Pair Plot

**shows all the plots for the data**
- it smartly selects all the numerical data from the dataset and plot's the data
- if same column's being plot with itslf, it shows a histogram
- if different column's being plot with different columns, it shows a scatter plot.

```python
sns.pairplot(data = iris_df)
```
