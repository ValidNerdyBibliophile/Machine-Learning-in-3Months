```python
#9. Concatenate: Create two data frames using the following two Dictionaries,
# Concatenate those two data frames and create a key for each data frame.

german_cars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
g_cars = pd.DataFrame.from_dict(german_cars)

japanese_cars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
j_cars = pd.DataFrame.from_dict(japanese_cars)

cars = pd.concat([g_cars, j_cars], keys=["Germany", "Japan"])
cars
```

```python
#10. Create two data frames using the following two Dictionaries, Merge two data frames.
# append second data frame as a new column to the first data frame.

germany_cars = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
german_cars = pd.DataFrame.from_dict(germany_cars)
japanese_cars = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Sales': [141, 444, 1821 , 1620]}
japan_cars = pd.DataFrame.from_dict(japanese_cars)
merged_data = pd.merge(german_cars, japan_cars, on="Company")
merged_data
```
