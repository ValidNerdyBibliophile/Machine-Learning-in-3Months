```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("humidity.csv",index_col='datetime',parse_dates=['datetime'])
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df=df.iloc[:,0:4]
df
```

```python
df.plot(subplots=True,figsize=(10,10))
```

```python
df_month = df.resample("M").mean()
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(df_month['2013'].index, df_month.loc['2013', "Portland"], width=25, align='center')
```

```python
df_month['Portland'].plot(figsize=(8, 6))
```

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 11,9
decomposition = sm.tsa.seasonal_decompose(df_month['Portland'],model='Additive')
fig = decomposition.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df["Portland"])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
#As we can see p-value <=0.05 therefore the model is stationary, 
#no need to do further procedure. so we can diretly move to Moving AVG.
```

```python
df1=df.head(100)
df1
```

```python
df1["SMA4"]=df1["Portland"].rolling(window=4).mean()
df1
```

```python
fig=plt.figure(figsize=(15,10))
plt.plot(df1["Portland"],label="Original")
plt.plot(df1["SMA4"],label="SMA4")
```

```python
df1["EMA"]=df1["Portland"].ewm(com=0.5,adjust=True).mean()
df1
```

```python
plt.plot(df1["Portland"],label="Portland_Humidity")
plt.plot(df1["EMA"],label="EMA")
plt.legend()
```
