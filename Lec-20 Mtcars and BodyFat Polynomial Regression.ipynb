```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mtcars_data = pd.read_csv("Mtcars.csv")
```

```python
mtcars_data.rename(columns = {'Unnamed: 0':'model'}, inplace = True)
```

```python
mtcars_data.head()
```

```python
mtcars_data.columns
```

```python
sns.pairplot(mtcars_data)
```

```python
x = mtcars_data[["disp"]]
y=mtcars_data["mpg"]
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model = LinearRegression()
lin_model.fit(x,y)
ypred = lin_model.predict(x)
```

```python
lin_model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x ,ypred, color = "r")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_transform = PolynomialFeatures(degree = 3)
```

```python
x_poly = poly_transform.fit_transform(x)
```

```python
poly_model = LinearRegression()
```

```python
poly_model.fit(x_poly,y)
ypred = poly_model.predict(x_poly)
```

```python
pd.DataFrame({"Actual":y,"Predicted":ypred})
```

```python
poly_model.score(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,ypred,color = "r")
```
