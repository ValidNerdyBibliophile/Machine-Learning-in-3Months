```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
x = data.drop(columns = ["Outcome"])
```

```python
y = data["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression
```

#### GridSeachCV
- estimators: Algorithm
- tuning parameters: solver, penalty and  c
- cv: cross validation -> kfold
- scoring: accuracy

Best Accuracy-> Parameters ->Can be use to build our models

```python
help(LogisticRegression)
```

```python
model = LogisticRegression()
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
penalty = ["l1", "l2", "elasticnet"]
c = [100,10,1,0.01,0.1,0.001]
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

kfoldcv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state = 1)
```

```python
help(RepeatedStratifiedKFold)
```

```python
grid = dict(solver =  solver, penalty = penalty, C = c)
grid
```

```python
grid_cv = GridSearchCV(estimator = model,param_grid = grid, cv = kfoldcv, scoring = "accuracy")
```

```python
res = grid_cv.fit(x,y)
```

```python
res.best_score_
```

```python
res.best_params_
```
