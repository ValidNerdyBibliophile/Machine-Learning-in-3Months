```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### x and y should be of same size

```python
x = [1,3,5,7]
y = [2,4,8,6]
```

## Line Plot

```python
plt.plot(x,y)
```

##### Marker
Marker's usually show the intersection points
Markers that can be used are:
- o
- *
- ^
- +
- s #block or a square

```python
plt.plot(x,y,marker = "o")
```

### Color
- m
- b
- g
- r
- y
- k

```python
plt.plot(x,y,marker = "^",color = "k")
```

### Labels

```python
plt.plot(x,y,marker = "^",color = "k", label = "MyLine")
plt.legend()
```

```python
plt.plot(x,y,marker = "^",color = "m", label = "MyLine")
plt.plot(y,x,marker = "+",color = "g", label = "HerLine")
plt.legend()
```

```python
plt.plot(y,x,marker = "+",color = "b", label = "HerLine")
plt.legend()
plt.xlabel("My X Axis")
plt.ylabel("My Y Axis")
```

### Title

```python
plt.plot(y,x,marker = "+",color = "y", label = "HerLine")
plt.legend()
plt.title("My First Plot")
```

### Line Style

```python
plt.plot(y,x,marker = "+",color = "r", label = "HerLine",linestyle = "--")
```

```python
plt.plot(y,x,marker = "s",color = "y", label = "HerLine",linestyle = "--",markerfacecolor = "b")
```

### Ticks/ Manipulate the axis 

```python
plt.plot(x,y,marker = "s",color = "y", label = "HerLine",linestyle = "--",markerfacecolor = "b")
plt.xticks([2,4,6])
plt.yticks([1,3,5])
```

### Positions

```python
plt.plot(x,y,marker = "s",color = "k", label = "MyLine",linestyle = "--",markerfacecolor = "r")
plt.legend(loc = "lower center")
plt.title("My First Plot")
plt.xlabel("My X")
plt.ylabel("My Y")
plt.xticks([2,4,6],["a","b","c"])
plt.yticks([1,3,5])
```

## Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s =100)  # s = size of the points
plt.show()
```

```python
plt.scatter(x,y, c =[0,1,0,1])
plt.show()
```

## Bar Chart

```python
plt.bar(x,y)
```

```python
plt.bar(x,y,color = ["r","g","b","k"])
```
