### Apriori Algorithm
Takes Frequent Item set to get the Association Rules.

Let's say, we went to DMART. In DMART, let's say, where there's a compartment of breads, there is also compartments of jam, egg, cheese spread etc. So what we basically see is, with bread what all are the other products that we can sell with bread. This increases Sales and Business of the vendors.

Suppose on Amazon we are searching for a mobile. We also get a list of frequently bought together products, like a mobile covers, earphones, tempered glass or accesories. bought together.

The relation between Laptop - Laptop Bag, Mobile - Mobile Cover, Bread - Jam, Egg's is calculated by Apriori Algorithm where we check how strong the association rule is between the frequent item sets, helping the vendor to improve their business and grow their business, Gives insigts to build strategies.

#### Terms Related to the Apriori Algorithm

- ***Support***
    - Popularity of that particular product/item or the ocurence of the product. 
    - Suppose we have a product A,Support of the product A is nothing but the ratio of frequency of A to the total number of transaction. ie, in that transaction how many times that product is been occured.

- ***Confidence***
    - Confidence is the likelihood where whenever an user purchases prod A will also purchases product B.
    - How many users bought milk and bread together. The probability of those products is called Confidence. 
    ***A -> B = freq(A,B) / freq(A)***
    
    ***ie. No of transactions having A and B / No. of transactions of A***
    
- ***Lift***
    - Increase in the sale of A, when you sell B.
    
   ***A -> B = confidence(A,B) / Suppor(B)***
   - If:
       - lift(A -> B) = 1:  No correlation ie Product are not related to each other
       - lift (A -> B) > 1: Positive Correlation ie bought together
       - lift (A -> B) < 1:  Negative Correlation ie not bought together

#### Steps:
1. Genereate Frequent Item Set
2. Rule Genereation

#### Steps of Apriori Algorithm
- Give Minimum Support and Minimum Confidence
- Calculate the support for all transactions and we pick the items in the transaction with higher or equal minimum support and discard which are not.
- We find rules for these picked subsets with higher confidence.
- Sort the rules as the decreasing order of lift.

#### with list comprehension

```python
trans = []
for i in range(df.shape[0]):
    trans.append([ str(df.values[i,j]) for j in range(df.shape[1]) if str(df.values[i,j])!="0"])
trans
```

```python
from apyori import apriori
rules = apriori(trans, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length =2)
```

```python
rules
```

```python
df_rules = pd.DataFrame(list(rules))
```

```python
df_rules.head()
```

```python
df_rules.iloc[0,2][0][0]
```

```python
for i in range(10):
    order = df_rules["ordered_statistics"][i][0]
    print(order[0])
    print(order[1])
    print(order[2])
    print(order[3])
```

```python
lhs = []
rhs = []
conf = []
lift = []
for i in range(df_rules.shape[0]):
    order = df_rules["ordered_statistics"][i][0]
    lhs.append(list(order[0]))
    rhs.append(list(order[1]))
    conf.append(order[2])
    lift.append(order[3])
```

```python
final_report = pd.DataFrame({"LHS":lhs, "RHS":rhs,"Support":df_rules.support, "Confidence": conf, "Lift":lift})
```
